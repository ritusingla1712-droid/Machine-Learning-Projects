```python
data.drop(["Address", "Date", "Postcode", "YearBuilt", "Lattitude", "Longtitude"],axis = 1, inplace = True)
```

```python
data[['Distance','Bedroom2', 'Bathroom', 'Car','Propertycount']] = data[['Distance','Bedroom2', 'Bathroom', 'Car','Propertycount']].fillna(0)
```

```python
data['Landsize'] = data['Landsize'].fillna(data.Landsize.mean())
data['BuildingArea'] = data['BuildingArea'].fillna(data.BuildingArea.mean())
```

```python
data = pd.get_dummies(data, drop_first = True)
```

```python
x = data.drop('Price', axis = 1)
y = data['Price']
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#training accuracy
model.score(x_train, y_train)*100
```

```python
#testing Accuracy
model.score(x_test,y_test)*100
```

## Ridge Regression 

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg= Ridge(alpha=50, max_iter=100,)
ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg.score(x_train, y_train)*100
```

```python
ridge_reg.score(x_test, y_test)*100
```

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
```

```python
y_predicted = ridge_reg.predict(x_test)
y_predicted
```

```python
r2_score(y_test,y_predicted)*100
```

```python
mean_squared_error(y_test,y_predicted)
```

```python
mean_absolute_error(y_test,y_predicted)
```
